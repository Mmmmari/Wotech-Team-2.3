# Coin Toss Game analysis

```python
import numpy as np

coin = ['Heads', 'Tails'] # 1D object(list), expected 1D array
number_of_throws = 1000 # size of the random list to generate

# Random coin toss result from 1000 throws
tosses = np.random.choice(coin, number_of_throws)
print(f'The resoults of tossing a coin 1000 times {tosses}')

# The sum of each opetion - Heads/Tails
total_head_tosses = np.sum(tosses == 'Heads') # count the number of heads
print(f'Total head count is {total_head_tosses}')
total_tail_tosses = np.sum(tosses == 'Tails') # count the number of heads
print(f'Total tail count is {total_tail_tosses}')
```

```python
# Probability of getting Heads vs Tails
probability_heads = total_head_tosses / number_of_throws
print(f'The probability of getting heads is {probability_heads}')

probability_tails = total_tail_tosses / number_of_throws
print(f'The probability of getting tails is {probability_tails}')
```

### Uniform distribution




```python
coordinates2 = np.random.rand(100000, 2)

x2 = coordinates2[:, 0]
y2 = coordinates2[:, 1]

plot = plt.scatter(x2, y2, s=1, color='green')
plt.title('100000 Randomly selected coordinates')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show(plot)

# all the values have an equal probability to be selected, therefore it shows uniform distribution
```

# TEAMWORK

Design a simulation where you toss two coins simultaneously. Record the possible outcomes (heads or tails for each coin) and run the simulation for multiple trials (e.g., 100, 500, or 1000 tosses). After collecting the data:


Calculate the probabilities of each outcome (HH, HT, TH, TT).

Visualize the results using a bar chart or pie chart to represent the frequencies and probabilities of each outcome.

Analyze whether the results align with the expected theoretical probabilities.

```python
# Heads == 0, Tails == 1

# 100 tosses
coin1_toss_100 = np.random.randint(0, 2, 100)
coin2_toss_100 = np.random.randint(0, 2, 100)

# combine the two coins
combined_coins_100 = np.column_stack((coin1_toss_100, coin2_toss_100))

result_count_100 = {'HH': 0, 'HT': 0, 'TH': 0, 'TT':0}

# Count outcomes
for result in combined_coins_100:
    if tuple(result) == (0, 0):
        result_count_100['HH'] += 1
    elif tuple(result) == (0, 1):
        result_count_100['HT'] += 1
    elif tuple(result) == (1, 0):
        result_count_100['TH'] += 1
    elif tuple(result) == (1, 1):
        result_count_100['TT'] += 1

print(result_count_100)

# probabilities
prob_HH_100 = result_count_100['HH']/100
print(f'After 100 throws the probability to get heads/heads is {prob_HH_100}.')
prob_HT_100 = result_count_100['HT']/100
print(f'After 100 throws the probability to get heads/tails is {prob_HT_100}.')
prob_TH_100 = result_count_100['TH']/100
print(f'After 100 throws the probability to get tails/heads is {prob_TH_100}.')
prob_TT_100 = result_count_100['TT']/100
print(f'After 100 throws the probability to get tails/tails is {prob_TT_100}.')

```

```python
# probability bar chart
prob_100 = [prob_HH_100, prob_HT_100, prob_TH_100, prob_TT_100]
prob_500 = [prob_HH_500, prob_HT_500, prob_TH_500, prob_TT_500]
prob_1000 = [prob_HH_1000, prob_HT_1000, prob_TH_1000, prob_TT_1000]

# Define the x-axis locations
x = np.arange(len(result_count_1000))

# Plotting
fig, ax = plt.subplots()

# Bar width
width = 0.35

# Stacked bars
bar1 = ax.bar(x, prob_100, width, label='100 tosses')
bar2 = ax.bar(x, prob_500, width, bottom=prob_100, label='500 tosses')
bar3 = ax.bar(x, prob_1000, width, bottom=np.array(prob_100) + np.array(prob_500), label='1000 tosses')

# Labels and title
ax.set_xlabel('Outcomes')
ax.set_ylabel('Probabilities')
ax.set_title('Stacked Bar Chart of Coin Toss Outcomes Probabilities')
ax.set_xticks(x)
ax.set_xticklabels(result_count_1000)
ax.legend()

# Show the plot
plt.show()
```
