```python
import pandas as pd
import matplotlib.pyplot as plt

pokemon_df = pd.read_csv('/content/Pokemon.csv')
pokemon_df.head(5)

# Create a hist of column speed
plt.hist(pokemon_df['Speed'], color = 'maroon', edgecolor = 'thistle')
plt.xlabel('Total')
plt.ylabel('Frequency')

# add min, max and average lines
plt.axvline(pokemon_df["Speed"].mean(), color='grey', linestyle='dashdot', linewidth=1)
plt.axvline(pokemon_df["Speed"].max(), color='red', linestyle='dashed', linewidth=1)
plt.axvline(pokemon_df["Speed"].min(), color='blue', linestyle='dashed', linewidth=1)
```

# CORRELATION

```python
# scatter plot between Attack and Defense variables
plt.scatter(pokemon_df['Attack'], pokemon_df['Defense'], color='coral', alpha=0.5)
plt.xlabel('Attack')
plt.ylabel('Defense')

# scatter plot for Speed and Defense
plt.scatter(pokemon_df['Speed'], pokemon_df['Defense'], color='coral', alpha=0.5)
plt.xlabel('Speed')
plt.ylabel('Defense')
plt.title('Speed vs Defense')

# scatter plot HP and Defense
plt.scatter(pokemon_df['HP'], pokemon_df['Defense'], color='maroon', alpha=0.5)
plt.xlabel('HP')
plt.ylabel('Defense')
plt.title('HP vs Defense')
```

## Pearson Correlation

```python
# Result of this correlation calculation is between -1 and 1
# -1 perfect negative relationship
# 1 perfect positive relationship

# Pearson correlation between Attack and Defense
correlation1 = pokemon_df['Attack'].corr(pokemon_df['Defense'])
print(f'The correlation betwwen Attach and Defense is {correlation1}. That is moderately positive correlation.')

# Pearson correlation between Attack and Defense
correlation2 = pokemon_df['Speed'].corr(pokemon_df['Defense'])
print(f'The correlation betwwen Speed and Defense is {correlation2}. There is close to none correlation.')

#
# Select only numeric columns

numeric_columns = pokemon_df.select_dtypes(include=['number'])
numeric_columns = numeric_columns.drop('#', axis=1)
numeric_columns
target_column = 'Attack'
correlation_results = {} #key = column, val=pearson_correlation
for column in numeric_columns:
  #if column != target_column:
    correlation = pokemon_df[target_column].corr(pokemon_df[column])
    correlation_results[column] = correlation
correlation_results
```

## Correlation Matrix

### Visualising with seaborn

```python
import seaborn as sns

correlation_matrix = numeric_columns.corr() # all columns are numeric
correlation_matrix

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Spectral')
plt.title('Correlation Matrix')
```

## Linear regression (predict one value based on the other)

```python
sns.regplot(x='Attack', y='Defense', data=pokemon_df, color='purple', ci=99, marker='x', line_kws={'color':'red'})
plt.title('Attack vs Defense')
print(f'Pearson correlation coefficient: {correlation1}')
```

```python
sns.regplot(x='Speed', y='Defense', data=pokemon_df, color='purple', ci=99, marker='x', line_kws={'color':'red'})
plt.title('Speed vs Defense')
print(f'Pearson correlation coefficient: {correlation2}')
```

## Task: Visualizing the correlation of Attack and Defense variables of two Types: Grass and Water Type 1 Pokémon.

```python
# 1. Create two DataFrame Grass and Water
grass_df = pokemon_df[pokemon_df['Type 1'] == 'Grass']
water_df = pokemon_df[pokemon_df['Type 1'] == 'Water']

# 2. Create the regression plots for each (Grass and Water) &
# 3. Calculate the Pearson correlation for each DataFrame (variables: Attack and Defense)
sns.regplot(x='Attack', y='Defense', data=grass_df, color='green', marker='x', line_kws={'color':'red'})
plt.title('Grass Attack vs Defense')
print(f'Pearson correlation coefficient: {grass_df["Attack"].corr(grass_df["Defense"])}')
```

```python
sns.regplot(x='Attack', y='Defense', data=water_df, color='navy', marker='x', line_kws={'color':'red'})
plt.title('Water Attack vs Defense')
print(f'Pearson correlation coefficient: {water_df["Attack"].corr(water_df["Defense"])}')
```
