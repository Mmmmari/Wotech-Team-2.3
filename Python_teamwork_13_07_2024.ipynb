```python
# EASY

import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('/content/transaction_dataset.csv')
dataf = pd.DataFrame(dataset)

dataset['Gender'] = dataset['Gender'].astype(str)
dataset['Gender'].value_counts()
gender_counts = dataset['Gender'].value_counts(dropna=False).reset_index()
gender_counts.columns = ['Gender', 'Counts']
df = pd.DataFrame(gender_counts)

plt.bar(df['Gender'], df['Counts'], color=['thistle','skyblue', 'orange'], edgecolor='maroon')

plt.title('Customers per Gender')
plt.xlabel('Gender')
plt.ylabel('Amount')

```

```python
# MEDIUM

name_df = pd.DataFrame(dataset['Name'].value_counts().reset_index())
name_df.columns = ['Name', 'Counts']

filtered_names = name_df.iloc[:5]

plt.barh(filtered_names['Name'], filtered_names['Counts'], color='skyblue')
plt.title('Top 5 names')
plt.ylabel('Names')
plt.xlabel('Frequency')

plt.show()
```
