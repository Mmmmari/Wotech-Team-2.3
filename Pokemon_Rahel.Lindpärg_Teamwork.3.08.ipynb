```python
# required libraries

import pandas as pd
import matplotlib.pyplot as plt

pokemon_df = pd.read_csv('/content/Pokemon.csv')
pokemon_df.head(n=10)
```

```python
pokemon_df.sample(5) #random 5
```

```python
#How many rows and columns are in this data frame
pokemon_df.shape #first is rows another columns
```

```python
rows, columns = pokemon_df.shape #gives them separately (variables)
print(f'There are {rows} rows and {columns} columns')
```

```python
#How many unique names are in this table?
pokemon_df['Name'].nunique()
```

```python
#this is how Roberta did it
unique_names = pokemon_df['Name'].unique()
print(f'There are {len(unique_names)} unique names')
```

```python
#How many pokemons have Attack greater than 100?
attack_greater_than_100 = pokemon_df[pokemon_df['Attack'] > 100]
print(len(attack_greater_than_100))
```

```python
defense_greater_than_100 = pokemon_df[pokemon_df['Defense'] > 100]
print(len(defense_greater_than_100))
```

```python
#How many Pokemons have both (Attack and Defense more than 100)?
attack_defense_greater_than_100 = pokemon_df[(pokemon_df['Attack'] > 100) & (pokemon_df['Defense'] > 100)]
print(len(attack_defense_greater_than_100))
```

```python
#Value counts method. Create a grouped DataFrame that counts the number of Pokemons in each group for Type 1.

type1_pokemons = pokemon_df['Type 1'].value_counts().reset_index()
print(type1_pokemons)
```

```python
#How to make a bar chart!
type1_pokemons.plot (x='Type 1', y='count', kind='bar', color= 'pink', edgecolor = 'black')
plt.title('Pokemons by Type 1')
plt.xlabel('Type 1')
plt.ylabel('Count')
plt.show()
```

```python
#Which pokemon has the biggest attack?
pokemon_df['Attack'].max() #the highest value of the attack
pokemon_df['Attack'].idxmax() #index of the highest value

highest_attack = pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmax()] #iloc=index location
print('Pokemon with the highest Attack is:',highest_attack)
```

```python
#Which has the lowest attack?
pokemon_df['Attack'].min()
pokemon_df['Attack'].idxmin()

lowest_attack = pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmin()]
print('Pokemon with the lowest Attack is:',lowest_attack)
```

```python
#adding a column
pokemon_df['My unique column'] = 'Pokemon'
pokemon_df.head()
```

```python
#adding calculated column
pokemon_df['Attack_Defense'] =  pokemon_df['Attack'] + pokemon_df['Defense']
pokemon_df.head()
```

```python
# quick math calculations
pokemon_df['Attack'].sum()
pokemon_df['Attack'].min()
pokemon_df['Attack'].max()
pokemon_df['Attack'].mean() #the average
pokemon_df['Attack above average'] = pokemon_df['Attack'] > pokemon_df['Attack'].mean()
pokemon_df.head()
```

```python
#Removing my unique column
pokemon_df = pokemon_df.drop(columns=['My unique column'])
pokemon_df.head()
```

```python
#HISTOGRAM (shows distribution)

# analyze total fo each pokemon

plt.hist(pokemon_df['Total'], color='green', edgecolor='blue', alpha=0.7) #alpha is transparenct) #HP+Attack+Defense
plt.xlabel('Total')
plt.ylabel('Number of Pokemons')
plt.axvline(pokemon_df['Total'].mean(), color='red', linestyle='dashed', linewidth=2)

#min line
plt.axvline(pokemon_df['Total'].min(), linestyle='dashdot', color='red')

#max line
plt.axvline(pokemon_df['Total'].max(), linestyle='dashdot', color='blue')

plt.grid(linestyle='--', alpha=0.7)



```

```python
#Make another but with less bins
value = plt.hist(pokemon_df['Total'], bins = 3, color='pink', edgecolor='black', alpha=0.7)

hist_values = value[0]
bin_edges = value[1]

print('Histogram counts:', hist_values)
print('Bin edges', bin_edges)
```

```python
#value_counts How we can calculate the same histogram with value counts?

pokemon_df['Total'].value_counts(bins=[179, 379, 579, 781])

(pokemon_df['Total']>579).sum()
```

```python
pokemon_df.hist(column='Total', by='Legendary', bins=3)
```

```python
# How many Pokémons are with 'Type 1'? == Water as a % pf Total?
water_pokemons = pokemon_df[pokemon_df['Type 1'] == 'Water']

# Calculate how many percent is Water from Total?
import numpy as np
water_percent = np.round((len(water_pokemons) / len(pokemon_df)) * 100, 2)
print(f'Water pokemons are {water_percent}% of total')

# What is the maximum Speed value? What is the minimum Speed value? What is the difference between max and min Speed?
max_speed = pokemon_df['Speed'].max()
min_speed = pokemon_df['Speed'].min()
speed_difference = max_speed - min_speed
print(f'Max speed is {max_speed}, min speed is {min_speed} and speed difference is {speed_difference}')

#Filter DataFrame to include only the Pokemon with Speed bigger or equal to 80.
speed_greater_than_80 = pokemon_df[pokemon_df['Speed'] >= 80]
#How many Pokemon met this criterion (bigger than or equal to)
print(len(speed_greater_than_80))
#Make a histogram to display this
speed_greater_than_80.hist(column='Speed', bins=5)








```
