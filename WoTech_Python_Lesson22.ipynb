# RANDOM WALK

```python
n = 100000 # defrining the number of steps

# initialize x and y
x = np.zeros(n)
y = np.zeros(n)

import random
val = random.randint(1, 4) # generate random int between 1 and 4

# use a for loop to modify the array based on random walk outcome
for i in range(1, n):
    val = random.randint(1, 4) # to get a direction of the step (4 directions in total)

    if val == 1: # dot moves x coordinate +1
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]

    elif val == 2: # dot moves x coordinate -1
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]

    elif val == 3: # dot moves y coordinate +1
        x[i] = x[i - 1]
        y[i] = y[i - 1] + 1

    else: # dot moves y coordinate -1
        x[i] = x[i - 1]
        y[i] = y[i - 1] - 1

```

```python
import pylab

# use pylab to visualize the random walk
plt.figure(figsize=(10, 6))  # Increase figure size
plt.plot(x, y, color='maroon', lw=2, alpha=0.8)  # Line with transparency
plt.scatter(x[0], y[0], c='green', s=100, zorder=5, label='Start')  # Mark start point
plt.scatter(x[-1], y[-1], c='red', zorder=5, s=100, label='End')  # Mark end point
pylab.title('Random Walk')
plt.legend()
pylab.show()

# save the file as .png
plt.savefig('random_walk.png', dpi=300, bbox_inches='tight')
```

# TEAMWORK

```python
# Simulate a 1D simple random walk where at each step can be moving only up or down

n = 1000
position = [0]
up_steps = 0

for i in range(1, n):
    step = random.choice([-1, 1])
    position.append(position[-1] + step)

    if step == 1:
        up_steps += 1

plt.figure(figsize=(10, 6))
plt.plot(position, label='Position Over Time', color='maroon')
plt.title(f"1D Simple Random Walk - Total Up Steps: {up_steps}", fontsize=16)
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.legend()
plt.show()

print(f"Total number of up steps: {up_steps}")
```
